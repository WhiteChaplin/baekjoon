```python
#https://www.acmicpc.net/problem/2164
"""
N장의 카드가 있다. 각각의 카드는 차례로 1부터 N까지의 번호가 붙어 있으며, 1번 카드가 제일 위에, N번 카드가 제일 아래인 상태로 순서대로 카드가 놓여 있다.

이제 다음과 같은 동작을 카드가 한 장 남을 때까지 반복하게 된다. 우선, 제일 위에 있는 카드를 바닥에 버린다. 그 다음, 제일 위에 있는 카드를 제일 아래에 있는 카드 밑으로 옮긴다.

예를 들어 N=4인 경우를 생각해 보자. 카드는 제일 위에서부터 1234 의 순서로 놓여있다. 1을 버리면 234가 남는다. 여기서 2를 제일 아래로 옮기면 342가 된다.
3을 버리면 42가 되고, 4를 밑으로 옮기면 24가 된다. 마지막으로 2를 버리고 나면, 남는 카드는 4가 된다.

N이 주어졌을 때, 제일 마지막에 남게 되는 카드를 구하는 프로그램을 작성하시오.

1. 맨 위에 있는 카드를 꺼낸다.
2. 제일 위에 있는 카드를 pop()에서 0번 인덱스로 옮긴다.
3. 카드가 1장 남을 때 까지 1번 2번 계속 반복


문제가 있다 리스트 큐의 고질적인 문제인데 빅오가 O(1)이 아닌 O(n)이란다...그래서 실행 속도가 느렸던 것이다.
디큐가 실행속도가 더 빨라서 디큐로 하니 문제가 풀렸다.
디큐는 오히려 리스트 형식과 비슷한 거 같은데 append로 값을 추가하고 큐랑 달리 오른쪽에 데이터가 적재된다.
그리고 pop은 오른쪽에서 값을 꺼내고, popleft는 왼쪽에서 값을 꺼낸다.
여기서는 [1,2,3,4,5,6] 리스트를 주고 왼쪽 인덱스가 맨 위라고 설정한 후 풀었다.
다 매
"""
from collections import deque 
num = int(input())
num_list = deque()
for i in range(1,num+1):
    num_list.append(i)
while len(num_list) > 1:
    num_list.popleft()
    num_list.append(num_list.popleft())
print(num_list[0])
```
