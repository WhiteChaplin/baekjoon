```python
#https://www.acmicpc.net/problem/1463
"""
정수 X에 사용할 수 있는 연산은 다음과 같이 세 가지 이다.

X가 3으로 나누어 떨어지면, 3으로 나눈다.
X가 2로 나누어 떨어지면, 2로 나눈다.
1을 뺀다.
정수 N이 주어졌을 때, 위와 같은 연산 세 개를 적절히 사용해서 1을 만들려고 한다. 연산을 사용하는 횟수의 최솟값을 출력하시오.
점화식 : Sn = max(Sn-1, Sn-2+an) or Sn = min(Sn-1, Sn-2+an) 
"""

memo = {1:0, 2:1, 3:1}
number = int(input())
def make1(num):
    if num in memo:
        return memo[num]
    if num % 3 == 0 and num % 2 == 0 :
        memo[num] = min(make1(num//3), make1(num//2))+1
    elif num % 3 == 0 :
        memo[num] = min(make1(num//3), make1(num-1))+1
    elif num % 2 == 0 :
        memo[num] = min(make1(num//2), make1(num-1))+1
    else:
        memo[num] = make1(num-1)+1
    return memo[num]

print(make1(number))
```
